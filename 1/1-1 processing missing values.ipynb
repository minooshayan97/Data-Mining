```python
from sklearn.impute import SimpleImputer
```

```python
imputer1=SimpleImputer(strategy='mean')
data5 = df.copy()
data5.iloc[:,2:5]=imputer1.fit_transform(data5.iloc[:,2:5])
data5
```

```python
imputer2=SimpleImputer(strategy='median')
data6 = df.copy()
data6.iloc[:,2:5]=imputer2.fit_transform(data6.iloc[:,2:5])
data6
```

```python
from sklearn.impute import KNNImputer
```
