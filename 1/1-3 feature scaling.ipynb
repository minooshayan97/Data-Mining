```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('OUT 1-2 data(categorial values encoded).csv')
df
```

```python
from sklearn.preprocessing import MinMaxScaler,Normalizer,StandardScaler
```

```python
pop_growth=np.array(df['Population growth'])
pop_growth
```

```python
scale1 = MinMaxScaler(feature_range=(0, 1))
new_pop_growth = scale1.fit_transform(pop_growth[:, np.newaxis])
new_pop_growth
```

```python
population = np.array(df['Total population'])
population
```

```python
new_population = scale1.fit_transform(population[:, np.newaxis])
new_population
```

```python
area = np.array(df['Area (sq. km)'])
area
```

```python
new_area = scale1.fit_transform(area[:, np.newaxis])
new_area
```

```python
data1 = df.copy()
data1['Population growth'] = new_pop_growth
data1['Total population'] = new_population
data1['Area (sq. km)'] = new_area
data1
```

```python
scale2=StandardScaler()
new_pop_growth = scale2.fit_transform(pop_growth[:, np.newaxis])
new_pop_growth
```

```python
new_population = scale2.fit_transform(population[:, np.newaxis])
new_population
```

```python
new_area = scale2.fit_transform(area[:, np.newaxis])
new_area
```

```python
data2 = df.copy()
data2['Population growth'] = new_pop_growth
data2['Total population'] = new_population
data2['Area (sq. km)'] = new_area
data2
```

```python
scale3=Normalizer(norm='l2')
new_pop_growth = scale3.fit_transform([pop_growth])
new_pop_growth
```

```python
new_population = scale3.fit_transform([population])
new_population
```

```python
new_area = scale3.fit_transform([area])
new_area
```

```python
data3 = df.copy()
data3['Population growth'] = new_pop_growth.T
data3['Total population'] = new_population.T
data3['Area (sq. km)'] = new_area.T
data3
```

```python
data2.to_csv('OUT 1-3 data(standard-transformed).csv', index=False)
```
