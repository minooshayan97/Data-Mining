```python
import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('OUT 1-4 data(outlier_passed).csv')
df
```

```python
s = {'x':df['Total population'],'x2':df['Total population']**2}
x = pd.DataFrame(data=s)
x
```

```python
y = df['Coronavirus Cases']
y
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
coefs=pd.DataFrame(model.coef_,x.columns,columns=['coefficient'])
coefs
```
