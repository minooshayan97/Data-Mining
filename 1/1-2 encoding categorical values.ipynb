```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
df=pd.read_csv('OUT 1-1 data(processed missing values).csv')
df
```

```python
df.info()
```

```python
international_visitors=LabelEncoder()
df['International Visitors']=international_visitors.fit_transform(df['International Visitors'])
df
```

```python
encoder=OneHotEncoder()
encoded_international_visitors=pd.DataFrame(encoder.fit_transform(df[['International Visitors']]).toarray())
encoded_international_visitors.rename(columns={0:'A',1:'B',2:'C',3:'D'},inplace=True)
encoded_international_visitors
```

```python
new_df=df.join(encoded_international_visitors)
new_df.drop(['International Visitors','D'],inplace=True,axis=1)
new_df
```

```python
new_df.to_csv('OUT 1-2 data(categorial values encoded).csv', index=False)
```
